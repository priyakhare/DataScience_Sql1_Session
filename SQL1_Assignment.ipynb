```python
#*******************************************************************************************************

#   Task 2 : To Show average Hours per week for all men  working in private Sector 

#********************************************************************************************************
# importing the packages and giving alias, for sqldb , importing pandasql

import pandas as pd
import sqlite3
import re


# Creating an Adult DataFrame as in Program 1 
adultDf = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
ColumnNames= ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','(>50K,<=50K)']
adultDf.columns = ColumnNames

# REmoving the hyphen from the Column names to make them accessible in the query
adultDf.columns = [re.sub("[- ]", "", col) for col in adultDf.columns]

# Creating a connection to the sqladb DataBase and creating a Adult Table 

conn = sqlite3.connect("sqladb.db")
adultDf.to_sql("Adults",conn, if_exists="replace")

# To Show Average Hours per week for people working in private sector 

print("\n Average hours per week by men working in private working class are :\n",'-'*80)

sqlQuery = " SELECT round(avg(hoursperweek),2) as  AvgHoursPerWeek  , workclass FROM Adults GROUP BY workclass HAVING workclass LIKE '%Private%' AND sex like '%Male%';"
#print(sqlQuery)

print(pd.read_sql_query(sqlQuery, conn))
conn.close()

```

```python
#*******************************************************************************************************

#   Task 3 : To Show frequency Table for Education , Occupation and relationship separately 

#********************************************************************************************************
# importing the packages and giving alias, pandas, sqlite3,re (Regular expression)

import pandas as pd
import sqlite3
import re


# Creating an Adult DataFrame as in Program 1 
adultDf = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
ColumnNames= ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','(>50K,<=50K)']
adultDf.columns = ColumnNames

# REmoving the hyphen from the Column names to make them accessible in the query
adultDf.columns = [re.sub("[- ]", "", col) for col in adultDf.columns]

# Creating a connection to the sqladb DataBase and creating a Adult Table 
conn = sqlite3.connect("sqladb.db")
adultDf.to_sql("Adults",conn, if_exists="replace")

# To Show Frequency table for Education 

print('-'*80,"\nThe Frequency of people based on the Education is :\n",'-'*80)

sqlQuery = " SELECT COUNT(education)  as Frequency , education FROM Adults GROUP BY education  ;"
#print(sqlQuery)

print(pd.read_sql_query(sqlQuery, conn))

# To Show frequency Table based on Ocupation

print('-'*80,"\nThe Frequency of people based on the Occupation is :\n",'-'*80)
sqlQuery = " SELECT COUNT(occupation)  as Frequency , occupation FROM Adults GROUP BY occupation  ;"
#print(sqlQuery)

print(pd.read_sql_query(sqlQuery, conn))


# To Show frequency Table based on Relationship

print('-'*80,"\nThe Frequency of people based on the RelationShip is :\n",'-'*80)
sqlQuery = " SELECT COUNT(relationship)  as Frequency , relationship FROM Adults GROUP BY relationship  ;"
#print(sqlQuery)

print(pd.read_sql_query(sqlQuery, conn))

conn.close()

```
